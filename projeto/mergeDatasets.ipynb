```python
import pandas as pd
```

```python
df = pd.read_csv('data/filtered_data.csv')
df_journal_replace = pd.read_csv('data/journal_abrv_replace.csv', header=None, names=['nome_abreviado', 'nome_completo'])
df_sjr =  pd.read_csv('data/sjr.csv', delimiter=';')
```

```python
df['journal'] = df['journal'].str.lower()
df_journal_replace = df_journal_replace.dropna(subset=['nome_abreviado', 'nome_completo'])
```

```python
journal_dict = dict(zip(df_journal_replace['nome_abreviado'], df_journal_replace['nome_completo']))
df['journal'] = df['journal'].map(lambda x: journal_dict.get(x, x))
```

```python
# Salvar o resultado em um novo arquivo CSV
df.to_csv('full_name.csv', index=False)

```

```python
df_full = pd.read_csv('full_name.csv')
```

```python
data = df_full.copy()
data=data[["doi","pubmed_id",'journal','title','authors', 'publish_time', 'url']]
data['authors'] = data['authors'].astype(str)
df_sjr['Title'] = df_sjr['Title'].str.lower()
```

```python
df_merged = data.merge(df_sjr, left_on='journal', right_on='Title', how='inner')
df_merged.to_csv('df_merged.csv', index=False)
```

```python
df_merged['total_docs'] = df_merged['Total Docs. (2022)'] + df_merged['Total Docs. (3years)']
df_merged.rename(columns={
          'Total Refs.': 'total_refs', 'Total Cites (3years)': 'total_cits',
          'H index': 'h_index', 'Citable Docs. (3years)': 'citable_docs',
          'Cites / Doc. (2years)': 'avg_cit_per_doc', 'Ref. / Doc.': 'avg_ref_per_doc'}, inplace=True)
df_merged.columns

```

```python
df_final = df_merged[['doi', 'title', 'authors', 'journal', 'Rank', 'SJR', 'h_index', 'total_docs', 'total_refs', 'total_cits', 'avg_cit_per_doc', 'avg_ref_per_doc', 'Country', 'Region', 'Publisher', 'Coverage', 'Categories', 'Areas' ]]
df_final.to_csv('data/df_final.csv', index=False)
```
